```python
countries = df.loc[df.euros.nlargest(10).index, "pais_nombre"].to_list()

heatmap_values = []
for country in countries:
    longitud = round(df.loc[df.pais_nombre == country, "Longitude"].to_list()[0],2)
    latitud = round(df.loc[df.pais_nombre == country, "Latitude"].to_list()[0],2)
    euros = df.loc[df.pais_nombre == country, "euros"].to_list()[0]
    folium.Marker((latitud,longitud),
                  popup= f"{country}: {euros}  € en exportaciones",
                  tooltip= "Pincha aquí",
                  icon = folium.Icon(color = "red", icon = "dragon", prefix = "fa")).add_to(map_with_mini)
    heatmap_values.append([latitud,longitud,euros])
plugins.HeatMap(heatmap_values).add_to(map_with_mini)
display(map_with_mini)
```

```python
countries = df.loc[df.euros.nlargest(10).index, "pais_nombre"].to_list()

heatmap_values = []
def asignar_color(euros):
    if euros > 1000000000:  # Ejemplo de una cantidad para definir el color
        return 'green'
    elif euros > 500000000:
        return 'orange'
    else:
        return 'red'

for country in countries:
    longitud = round(df.loc[df.pais_nombre == country, "Longitude"].to_list()[0], 2)
    latitud = round(df.loc[df.pais_nombre == country, "Latitude"].to_list()[0], 2)
    euros = df.loc[df.pais_nombre == country, "euros"].to_list()[0]

    color = asignar_color(euros)

    folium.Marker(
        (latitud, longitud),
        popup=f"{country}: {euros} € en exportaciones",
        tooltip="Pincha aquí",
        icon=folium.Icon(color=color, icon='info-sign')
    ).add_to(map_with_mini)
    
#plugins.HeatMap(heatmap_values).add_to(map_with_mini)
display(map_with_mini)
```

```python
# Ordenar los países por importancia en euros
df_sorted = df.sort_values(by='euros', ascending=False)
top_countries = df_sorted['pais_nombre'].head(10).tolist()

# Crear los marcadores para cada país
for i, country in enumerate(top_countries, start=1):
    longitud = round(df.loc[df.pais_nombre == country, "Longitude"].iloc[0], 2)
    latitud = round(df.loc[df.pais_nombre == country, "Latitude"].iloc[0], 2)
    euros = df.loc[df.pais_nombre == country, "euros"].iloc[0]

    folium.Marker(
        (latitud, longitud),
        popup=f"Posición {i}: {country}: {euros} € en exportaciones",
        tooltip="Pincha aquí",
        icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(map_with_mini)
display(map_with_mini)
```

```python
countries = df.loc[df.euros.nlargest(10).index, "pais_nombre"].to_list()
for i, country in enumerate(top_countries, start=1):
    longitud = round(df.loc[df.pais_nombre == country, "Longitude"].iloc[0], 2)
    latitud = round(df.loc[df.pais_nombre == country, "Latitude"].iloc[0], 2)
    euros = df.loc[df.pais_nombre == country, "euros"].iloc[0]

    folium.Circle(
        location=(latitud, longitud),
        radius=euros / 1000000,  # Escalar el radio según la cantidad de exportaciones
        popup=f"Posición {i}: {country}: {euros} € en exportaciones",
        tooltip="Pincha aquí",
        color='blue',
        fill=True,
        fill_color='blue'
    ).add_to(map_with_mini)
    
    display(map_with_mini)
```

```python
import matplotlib.pyplot as plt

# Supongamos que tienes un DataFrame 'df_exportaciones' con datos de exportaciones por país y año

# Seleccionar países específicos para visualizar su evolución temporal
paises_seleccionados = ['España', 'Francia', 'Alemania', 'Italia']

for pais in paises_seleccionados:
    datos_pais = df[df['pais_nombre'] == pais]
    plt.plot(datos_pais['año'], datos_pais['cantidad_exportada'], label=pais)

plt.xlabel('Año')
plt.ylabel('Cantidad Exportada')
plt.title('Evolución Temporal de Exportaciones por País')
plt.legend()
plt.show()
```
