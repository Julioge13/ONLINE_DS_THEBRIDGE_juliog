```python
import pandas as pd
import glob
import re
```

```python
ruta_archivo = r'C:\Users\julio\Desktop\The_Bridge\Temario\EDA\Dataset\Banco_de_Espanya\IPC\series868898012sc.csv'

# Cargar el archivo en un DataFrame saltando las primeras 5 filas
df = pd.read_csv(ruta_archivo, sep=';')

df
```

```python
df = df[['PERIODO','VALOR']].copy()
df
```

```python
# Convertir la columna 'PERIODO' a tipo fecha
df['PERIODO'] = pd.to_datetime(df['PERIODO'], format='%YM%m')

# Mostrar solo el año y el mes en el formato deseado (por ejemplo, 'Año-Mes')
df['PERIODO'] = df['PERIODO'].dt.strftime('%Y-%m')

df
```

```python
# Convertir la columna 'VALOR' a tipo numérico (float)
df['VALOR'] = pd.to_numeric(df['VALOR'].str.replace(',', '.'), errors='coerce')

df
```

```python
# Convertir la columna 'PERIODO' a tipo fecha si aún no lo está
df['PERIODO'] = pd.to_datetime(df['PERIODO'])

# Crear una nueva columna 'AÑO' para extraer el año de la columna 'PERIODO'
df['AÑO'] = df['PERIODO'].dt.year

# Calcular el IPC anual promediando los valores mensuales por año
ipc_anual = df.groupby('AÑO')['VALOR'].mean().reset_index()

# Renombrar las columnas
ipc_anual.columns = ['AÑO', 'IPC_ANUAL']

ipc_anual
```
