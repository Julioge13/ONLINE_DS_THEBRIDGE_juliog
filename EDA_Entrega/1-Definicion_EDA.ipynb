```python
import pandas as pd
import glob

# Directorios que contienen los archivos CSV
dir_importaciones = r'C:\Users\julio\Bootcamp_DS\ONLINE_DS_THEBRIDGE_juliog\EDA_Entrega\Dataset\IMPORTACIONES'
dir_exportaciones = r'C:\Users\julio\Bootcamp_DS\ONLINE_DS_THEBRIDGE_juliog\EDA_Entrega\Dataset\EXPORTACIONES'

# Obtener la lista de todos los archivos CSV en el directorio de importaciones
archivos_importaciones = glob.glob(dir_importaciones + "/*.csv")
archivos_exportaciones = glob.glob(dir_exportaciones + "/*.csv")

# Lista para almacenar todos los DataFrames
lista_dataframes = []

# Leer cada archivo CSV de importaciones y almacenarlo en la lista
for archivo in archivos_importaciones:
    df = pd.read_csv(archivo, sep=';', encoding='ISO-8859-1')
    lista_dataframes.append(df)

# Leer cada archivo CSV de exportaciones y almacenarlo en la lista
for archivo in archivos_exportaciones:
    df = pd.read_csv(archivo, sep=';', encoding='ISO-8859-1')
    lista_dataframes.append(df)

# Concatenar todos los DataFrames en uno solo
df_final = pd.concat(lista_dataframes, ignore_index=True)

df_final.drop(['periodo_mes', 'Unnamed: 9'], axis=1, inplace=True)
```
